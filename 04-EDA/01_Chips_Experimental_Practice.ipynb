```python
chips['Prize_INR']=np.where(chips['Chips_Type']=='Lays',20,10)
chips.head()
```

```python
# chips.drop(0)
chips.drop(columns=['Prize_INR'])
# chips.drop(columns=['Prize ₹'], inplace=True) -> will remove the columns from data completely
```

```python
chips #-> still has 'Prize_INR' as inplace is false by defaults
```

```python
chips['Revenue']=chips['Unit_Sold']*chips['Prize_INR']
chips
```

```python
chips.groupby('Day')
```

```python
chips.set_index(['Day','Chips_Type','Unit_Sold','Prize_INR','Revenue'])
```

```python
chips['COC']=np.where(chips['Chips_Type']=="Lays",17,8)
#  Here COC means cost on company per bag of chips like a 20 rs chips cost 17 rs to the company
chips
```

```python
chips.drop(columns='Revenue',inplace=True)
```

```python
chips['Sales']=chips['Prize_INR']*chips['Unit_Sold']
chips
```

```python
chips['Profit']=(chips['Prize_INR']-chips['COC'])*chips['Unit_Sold']
chips
```

```python
chips.drop(columns=['Profit'],inplace=True)
```

```python
chips.groupby(chips['Chips_Type'])['Unit_Sold'].sum()
```

```python
chips.groupby(['Chips_Type']).agg({'Unit_Sold' : 'sum' , 'Sales' : 'sum'}).reset_index()
```

```python
pivot = chips.pivot(columns='Chips_Type',index='Day',values='Sales')
pivot
```

```python
chips['Yesterday_Sales']=chips['Sales'].shift(2)
chips['Yesterday_Sales'].fillna(0,inplace=True)
chips
```

```python
chips['%_change']=chips['Sales']/chips['Yesterday_Sales'] *100
chips
```

```python
chips['UnitSold_Rank']=chips['Unit_Sold'].rank(method='first',ascending=False)
chips.sort_values(['UnitSold_Rank'])
```

```python
chips['cumulative_sales']=chips['Sales'].cumsum()
chips
```

```python
# .rolling() for like 3 days
lays=chips[chips['Chips_Type']=="Lays"].copy()
lays['3day']=lays['Sales'].rolling(3).sum()
lays
#  lays dataframes have only one day while in chips same day repeasts 2 times
#  so if same thing used in chips it will not give the '3days' data but strat suming after 3 columns
```
