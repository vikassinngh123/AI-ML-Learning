### Tensor **transpose** (for changing shape)

```python
t_A=torch.tensor([[1,2],
                  [3,4],
                  [5,6]])
t_B=torch.tensor([[7,8],
                  [9,10],
                  [11,12]])
```

```python
print("t_A.matmul(t_B) will give an error since it does not follow matrix multipication rule\n")

try:
    t_A.mm(t_B)
except:
  print("RuntimeError: mat1 and mat2 shapes cannot be multiplied (3x2 and 3x2)")
```

```python
t_B.T
```

```python
print(f"Orignal shapes of t_B-> {t_B.shape}\n")
print(f"Transposed shapes of t_B-> {t_B.T.shape}")
```

```python
print(f"matrix multipication of t_A and t_B transpose\n {t_A.mm(t_B.T)}\n")

print(f"matrix multipication of t_A transpose and t_B\n {t_A.T.mm(t_B)}")
```

### Tensor Aggregration

```python
ten_1=torch.tensor([[12,54,23],
                [1,8,94],
                [34,56,21]])
ten_1
```

```python
print(f"max of tensor-> {ten_1.max()}\n")
print(f"min of tensor-> {ten_1.min()}\n")
print(f"mean of tensor-> {ten_1.type(torch.float32).mean()}\n")
print(f"sum of tensor-> {ten_1.sum()}\n")
```

#### Positional min and max

```python
print(f"max of tensor position-> {ten_1.argmax()}\n")
print(f"min of tensor position-> {ten_1.argmin()}\n")
```

```python
# When we call argmax() without specifying a dimension (dim),
# PyTorch flattens the tensor into a 1D sequence to find the position:
# Index 0: 12
# Index 1: 54
# Index 2: 23
# Index 3: 1
# Index 4: 8
# Index 5: 94 -> (Maximum Value)
# Index 6: 34
# Index 7: 56
# Index 8: 21
# Since 94 is the largest element and sits at index 5, ten_1.argmax() gives 5.
```

### Reshaping , Stacking , Squeezing and Unsqueezing

```python
x=torch.arange(1.,10.)
x,x.shape
```

```python
x_reshaped1=x.reshape(1,9)
x_reshaped1,x_reshaped1.shape
```

```python
x_reshaped2=x.reshape(9,1)
x_reshaped2,x_reshaped2.shape
```

```python
x_reshaped3=x.reshape(3,3)
x_reshaped3,x_reshaped3.shape
```

```python
# View
y=x.view(1,9)
y,y.shape
```

```python
y[:,2]=34
y,x
```

```python
# Stacking tensors

x_stacked0=torch.stack([x,x,x,x],dim=0)
x_stacked0
```

```python
x_stacked1=torch.stack([x,x,x,x],dim=1)
x_stacked1
```

```python
# squeeze and unsqueeze
x_unsqueeze0=x_stacked0.unsqueeze(dim=0)
x_unsqueeze1=x_stacked0.unsqueeze(dim=1)
x_unsqueeze2=x_stacked0.unsqueeze(dim=2)
print(f"{x_unsqueeze0} \n shape of this unsqueeze tensor {x_unsqueeze0.shape}\n")
print(f"{x_unsqueeze1} \n shape of this unsqueeze tensor {x_unsqueeze0.shape}\n")
print(f"{x_unsqueeze2} \n shape of this unsqueeze tensor {x_unsqueeze0.shape}\n")
```

```python
x_squeeze0=x_unsqueeze0.squeeze(dim=0)
x_squeeze1=x_unsqueeze1.squeeze(dim=1)
x_squeeze2=x_unsqueeze2.squeeze(dim=2)
print(f"shape of this squeeze0 tensor {x_squeeze0.shape}\n")
print(f"shape of this squeeze1 tensor {x_squeeze1.shape}\n")
print(f"shape of this squeeze2 tensor {x_squeeze2.shape}\n")

print(f"all the squeeze tensor of unsqueeze tensor just remove the dim that was added and x_squeeze0/1/2 are all same \n\n{x_squeeze0}")
```
