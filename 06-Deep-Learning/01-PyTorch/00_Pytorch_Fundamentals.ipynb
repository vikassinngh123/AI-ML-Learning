### Random Tensor

```python
random_tensor=torch.rand(2,3,5)
random_tensor
```

```python
print(f"Shape {random_tensor.shape}")
print(f"ndim {random_tensor.ndim}")
print(f"dtype {random_tensor.dtype}")
print(f"device {random_tensor.device}")
```

```python
random_image_tensor=torch.rand(3,224,224) # color channels(R,G,B) , height , weight
print(f"Shape-> {random_image_tensor.shape}")
print(f"ndim-> {random_image_tensor.ndim}")
print(f"dtype-> {random_image_tensor.dtype}")
print(f"device-> {random_image_tensor.device}")
```

### Zeroes and Ones tensor

```python
zeroes=torch.zeros(2,3,3)
zeroes
```

```python
ones=torch.ones(size=(2,3,3))
print(ones)
print(f"\nShape-> {ones.shape}")
print(f"ndim-> {ones.ndim}")
print(f"dtype-> {ones.dtype}")
```

```python
ones_int=ones.type(torch.int32)
print(ones_int)
print(f"\nShape-> {ones_int.shape}")
print(f"ndim-> {ones_int.ndim}")
print(f"dtype-> {ones_int.dtype}")
```
