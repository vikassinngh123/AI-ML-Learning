```python
n=int(input("enter number: "))
a=np.ones((n,n) , dtype=int)
c=0
d=1
e=n
while n>1:
    a[1+c:e-1,1+c:e-1]=d+1
    c+=1
    d+=1
    n-=2
    e-=1

print(a)

```
