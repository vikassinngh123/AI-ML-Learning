```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import mnist
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
(X_train, y_train), (_, _) = mnist.load_data()
```

```python
X = X_train[:10000].reshape(10000, 784)
y = y_train[:10000]
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
pca=PCA(n_components=0.95)
X_pca=pca.fit_transform(X_scaled)
```

```python
print(f"Original dimension count: {X_scaled.shape[1]}")
print(f"Reduced dimension count (for 95% variance): {pca.n_components_}")
```

```python
pca_2d=PCA(n_components=2)
X_pca_2d=pca_2d.fit_transform(X_scaled)

sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))

sns.scatterplot(x=X_pca_2d[:, 0], y=X_pca_2d[:, 1], hue=y, palette="Set1",alpha=0.6,legend='full')

plt.title("PCA Visualization (2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

sns.despine()
plt.show()
```

## TSNE

```python
from sklearn.manifold import TSNE

tnse=TSNE(n_components=2,random_state=42,max_iter=1000)
X_tnse=tnse.fit_transform(X_scaled)

sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))

sns.scatterplot(x=X_tnse[:, 0], y=X_tnse[:, 1], hue=y, palette="tab10",alpha=0.6,legend='full')

plt.title("TSNE Visualization (2D)")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")

plt.legend(title="Classes", loc="upper right")

sns.despine()
plt.show()
```
