## Work exp V/s pay

```python
work_df=df.dropna(subset=["WorkExp","ConvertedCompYearly"])

world_work_df=work_df[(work_df["ConvertedCompYearly"]>0) & (work_df["ConvertedCompYearly"]<1000000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<30)]
cleaned_world_work_df=world_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_world_work_df=cleaned_world_work_df.groupby("WorkExp")['ConvertedCompYearly'].mean().reset_index()

#USA
USA_work_df=work_df[(work_df["ConvertedCompYearly"]>0) & (work_df["ConvertedCompYearly"]<1000000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<30) & (work_df['Country'].isin(['United States of America']))]
cleaned_USA_work_df=USA_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_USA_work_df=cleaned_USA_work_df.groupby("WorkExp")['ConvertedCompYearly'].mean().reset_index()

#India
India_work_df=work_df[(work_df["ConvertedCompYearly"]>0) & (work_df["ConvertedCompYearly"]<500000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<30) & (work_df['Country'].isin(['India']))]
cleaned_India_work_df=India_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_India_work_df=cleaned_India_work_df.groupby("WorkExp")['ConvertedCompYearly'].mean().reset_index()

```

##### Visual

## **Key Insights: Salary vs. Work Experience**

* **The US Market Premium:** The US consistently leads in compensation across all experience levels due to its highly mature tech ecosystem and premium software engineering labor costs.

* **Global Trend Stability:** The 'World' trendline remains incredibly smooth because its massive global sample size completely averages out regional economic extremes.

* **India's High-Experience Volatility**: India’s sharp spikes and dips past 20 years are caused by low sample sizes and extreme outlier sensitivity, a direct byproduct of the country's post-2000s IT boom.

* **The 15-Year Growth Plateau:** Across all regions, aggressive salary growth driven by early-career job-hopping levels off after 15–20 years as developers reach senior management ceilings.


```python
from matplotlib.lines import lineStyles
from matplotlib import markers
plt.figure(figsize=(15,8))
sns.set_theme(style='whitegrid',context='talk')

# World
sns.lineplot(
             data=cleaned_world_work_df,
             x="WorkExp",
             y="ConvertedCompYearly",
             marker='o',
             markersize=10,
             markeredgecolor='black',
             markerfacecolor='white',
             markeredgewidth=2,
             linestyle='-.',
             color='red',
             linewidth=2,
             label='Avg Annual Salary(World)'
             )

# USA
sns.lineplot(
             data=cleaned_USA_work_df,
             x="WorkExp",
             y="ConvertedCompYearly",
             marker='o',
             markersize=10,
             markeredgecolor='black',
             markerfacecolor='white',
             markeredgewidth=2,
             linestyle='-.',
             color='green',
             linewidth=2,
             label='Avg Annual Salary(USA)'
             )

#India
sns.lineplot(
             data=cleaned_India_work_df,
             x="WorkExp",
             y="ConvertedCompYearly",
             marker='o',
             markersize=10,
             markeredgecolor='black',
             markerfacecolor='white',
             markeredgewidth=2,
             linestyle='-.',
             color='blue',
             linewidth=2,
             label='Avg Annual Salary(India)'
             )

plt.title("Salary as work exp increase",fontsize=16,fontweight='bold')
plt.xlabel("Work Exp",fontsize=12,fontweight='bold')
plt.ylabel("Salary",fontsize=12,fontweight='bold')


plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.grid(axis='x',visible=False)

plt.legend(bbox_to_anchor=(1.05,1) ,loc='upper left' ,frameon=False)

sns.despine()

plt.show()
```

## Average Salary of Popular Language & Database

### Language

```python
popular_list = [
    'Python', 'JavaScript', 'Java', 'C++', 'C#', 'C', 'TypeScript', 'Go',
    'SQL', 'PHP', 'Swift', 'Kotlin', 'Rust', 'R', 'Ruby', 'MATLAB', 'Dart',
    'PowerShell', 'Visual Basic', 'Julia'
]

c_df=df.dropna(subset=["ConvertedCompYearly"]).copy()
c_df=c_df[(c_df["ConvertedCompYearly"]>0) & (c_df["ConvertedCompYearly"]<1000000)]

exploded_df = c_df.assign(Language=c_df['LanguageHaveWorkedWith'].str.split(';')).explode('Language')

filtered_df = exploded_df[exploded_df['Language'].isin(popular_list)]

Lang_df = filtered_df.groupby('Language')['ConvertedCompYearly'].mean().reset_index()
Lang_df = Lang_df.rename(columns={'ConvertedCompYearly': 'AvgSalary'}).sort_values(by='AvgSalary', ascending=False)

Lang_df
```

#### Visual

```python
plt.figure(figsize=(18,9))
sns.set_theme(style='whitegrid',context='talk')

sns.barplot(
            data=Lang_df,
            x='AvgSalary',
            y='Language',
            hue='Language',
            palette='plasma',
            saturation=1,
            edgecolor='black',
            gap=0.15,
            linewidth=2
           )

plt.title("Average Salary of Popular Language",fontsize=18,fontweight='bold')
plt.xlabel("Salary",fontsize=14,fontweight='bold')
plt.ylabel("Language",fontsize=14,fontweight='bold')

sns.despine()

plt.show()
```

### Database

```python
popular_databases = [
    'PostgreSQL', 'MySQL', 'SQLite', 'MongoDB', 'Redis',
    'Microsoft SQL Server', 'Amazon DynamoDB', 'MariaDB',
    'Elasticsearch', 'Firebase Realtime Database','Oracle'
]

database_df=df.dropna(subset=["ConvertedCompYearly","DatabaseHaveWorkedWith"]).copy()
database_df=database_df[(database_df["ConvertedCompYearly"]>0) & (database_df["ConvertedCompYearly"]<1000000)]
database_df=database_df[["DatabaseHaveWorkedWith","ConvertedCompYearly"]]

database_df=database_df.assign(Database=database_df['DatabaseHaveWorkedWith'].str.split(';')).explode('Database')

database_df=database_df[database_df['Database'].isin(popular_databases)]
database_df=database_df.groupby("Database")['ConvertedCompYearly'].mean().reset_index()
database_df=database_df.sort_values(by='ConvertedCompYearly',ascending=False)

database_df
```

##### Visual

```python
plt.figure(figsize=(16,8))
sns.set_theme(style='whitegrid',context='talk')

sns.barplot(
            data=database_df,
            x='ConvertedCompYearly',
            y='Database',
            hue='Database',
            palette='viridis',
            saturation=1,
            edgecolor='black',
            linewidth=2,
            gap=0.15
)

plt.title("Average Salary of Popular Database",fontsize=20,fontweight='bold')
plt.xlabel("Salary",fontsize=16,fontweight='bold')
plt.ylabel("Database",fontsize=16,fontweight='bold')

sns.despine()

plt.show()
```
