## Work exp V/s pay

```python
from numpy._core.numeric import full
work_df=df.dropna(subset=["WorkExp","ConvertedCompYearly"])

world_work_df=work_df[(work_df["ConvertedCompYearly"]>=5000) & (work_df["ConvertedCompYearly"]<=800000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<=25)]
cleaned_world_work_df=world_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_world_work_df=cleaned_world_work_df.groupby("WorkExp").agg({'ConvertedCompYearly':'mean'}).reset_index()
cleaned_world_work_df=cleaned_world_work_df.rename(columns={'ConvertedCompYearly':'World_AvgSalary'})

#USA
USA_work_df=work_df[(work_df["ConvertedCompYearly"]>=5000) & (work_df["ConvertedCompYearly"]<=800000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<=25) & (work_df['Country'].isin(['United States of America']))]
cleaned_USA_work_df=USA_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_USA_work_df=cleaned_USA_work_df.groupby("WorkExp").agg({'ConvertedCompYearly':'mean'}).reset_index()
cleaned_USA_work_df=cleaned_USA_work_df.rename(columns={'ConvertedCompYearly':'USA_AvgSalary'})

#India
India_work_df=work_df[(work_df["ConvertedCompYearly"]>=5000) & (work_df["ConvertedCompYearly"]<=100000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<=25) & (work_df['Country'].isin(['India']))]
cleaned_India_work_df=India_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_India_work_df=cleaned_India_work_df.groupby("WorkExp").agg({'ConvertedCompYearly':'mean'}).reset_index()
cleaned_India_work_df=cleaned_India_work_df.rename(columns={'ConvertedCompYearly':'India_AvgSalary'})

full_workexp_df=cleaned_India_work_df.merge(cleaned_world_work_df,on=['WorkExp'],how='outer')
full_workexp_df=full_workexp_df.merge(cleaned_USA_work_df,on=['WorkExp'],how='outer')
full_workexp_df
```

##### Visual

## **Key Insights: Salary vs. Work Experience**

* **The US Market Premium:** The US consistently leads in compensation across all experience levels due to its highly mature tech ecosystem and premium software engineering labor costs.

* **Global Trend Stability:** The 'World' trendline remains incredibly smooth because its massive global sample size completely averages out regional economic extremes.

* **India's High-Experience Volatility**: India’s sharp spikes and dips past 20 years are caused by low sample sizes and extreme outlier sensitivity, a direct byproduct of the country's post-2000s IT boom.

* **The 15-Year Growth Plateau:** Across all regions, aggressive salary growth driven by early-career job-hopping levels off after 15–20 years as developers reach senior management ceilings.


```python
graph_workexp=px.line(
                      full_workexp_df,
                      x='WorkExp',
                      y=['World_AvgSalary','USA_AvgSalary','India_AvgSalary'],
                      title='Salary as work exp increase',
                      template='plotly_dark',
                      labels={'value':'Salary','WorkExp':'WorkExp'},
                      markers=True,
                      color_discrete_sequence=['red','green','blue']
)

graph_workexp.update_layout(xaxis_title='WorkExp',yaxis_title='AvgSalary')
graph_workexp.show()

```

## Average Salary of Popular Language & Database

### Language

```python
popular_list = [
    'Python', 'JavaScript', 'Java', 'C++', 'C#', 'C', 'TypeScript', 'Go',
    'SQL', 'PHP', 'Swift', 'Kotlin', 'Rust', 'R', 'Ruby', 'MATLAB', 'Dart',
    'PowerShell', 'Visual Basic', 'Julia'
]

c_df=df.dropna(subset=["ConvertedCompYearly"]).copy()
c_df=c_df[(c_df["ConvertedCompYearly"]>=4000) & (c_df["ConvertedCompYearly"]<=800000)]

exploded_df = c_df.assign(Language=c_df['LanguageHaveWorkedWith'].str.split(';')).explode('Language')

filtered_df = exploded_df[exploded_df['Language'].isin(popular_list)]

Lang_df = filtered_df.groupby('Language')['ConvertedCompYearly'].mean().reset_index()
Lang_df = Lang_df.rename(columns={'ConvertedCompYearly': 'AvgSalary'}).sort_values(by='AvgSalary', ascending=False)

Lang_df.head()
```

#### Interactive Graph

```python
graph=px.bar(
             Lang_df,
             x='AvgSalary',
             y='Language',
             orientation='h',
             color='AvgSalary',
             color_continuous_scale='viridis',
             title='Average Salary of Popular Language',
             template='plotly_dark'
            )

graph.update_yaxes(categoryorder='total ascending')
graph.update_layout(coloraxis_showscale=False)
graph.show()
```

## Language Popularity V/s Avgsalary

```python
lang_stats = filtered_df.groupby('Language').agg(
    AvgSalary=('ConvertedCompYearly', 'mean'),
    DeveloperCount=('Language', 'count')
).reset_index()


fig_scatter = px.scatter(
    lang_stats,
    x='DeveloperCount',
    y='AvgSalary',
    text='Language', # Puts the name right next to the dot
    size='DeveloperCount', # Bigger dots for more popular languages
    color='AvgSalary',
    color_continuous_scale='plasma',
    title='Market Dynamics: Language Popularity vs. Average Salary',
    template='plotly_dark'
)


fig_scatter.update_traces(textposition='top center')
fig_scatter.update_layout(coloraxis_showscale=False)
fig_scatter.show()
```

### Database

```python
popular_databases = [
    'PostgreSQL', 'MySQL', 'SQLite', 'MongoDB', 'Redis',
    'Microsoft SQL Server', 'Amazon DynamoDB', 'MariaDB',
    'Elasticsearch', 'Firebase Realtime Database','Oracle'
]

database_df=df.dropna(subset=["ConvertedCompYearly","DatabaseHaveWorkedWith"]).copy()
database_df=database_df[(database_df["ConvertedCompYearly"]>=5000) & (database_df["ConvertedCompYearly"]<=800000)]
database_df=database_df[["DatabaseHaveWorkedWith","ConvertedCompYearly"]]

database_df=database_df.assign(Database=database_df['DatabaseHaveWorkedWith'].str.split(';')).explode('Database')

database_df=database_df[database_df['Database'].isin(popular_databases)]
database_df=database_df.groupby("Database")['ConvertedCompYearly'].mean().reset_index()
database_df=database_df.sort_values(by='ConvertedCompYearly',ascending=False)

database_df
```

#### Interactive Graph

```python
graph_db=px.bar(
                database_df,
                x='ConvertedCompYearly',
                y='Database',
                orientation='h',
                color='Database',
                title='Average Salary of Popular Database',
                template='plotly_dark'
              )

graph_db.update_layout(xaxis_title='AvgSalary',coloraxis_showscale=False)
graph_db.show()
```

## EDlevel vs Salary

```python
ed_levels_list = df['EdLevel'].dropna().unique().tolist()
ed_clean_map = {
    "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)": "Master's",
    "Bachelor’s degree (B.A., B.S., B.Eng., etc.)": "Bachelor's",
    "Associate degree (A.A., A.S., etc.)": "Associate",
    "Some college/university study without earning a degree": "Some College",
    "Professional degree (JD, MD, Ph.D, Ed.D, etc.)": "Professional/PhD",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "High School",
    "Primary/elementary school": "Primary",
    "Other (please specify):": "Other"
}

ed_df=df.dropna(subset=["ConvertedCompYearly","EdLevel",'WorkExp']).copy()
ed_df=ed_df[(ed_df["ConvertedCompYearly"] >= 5000) & (ed_df["ConvertedCompYearly"]<=800000) & (ed_df['EdLevel'].isin(ed_levels_list))]
ed_df=ed_df[["EdLevel","ConvertedCompYearly",'WorkExp']]
ed_df['EdLevel']=ed_df['EdLevel'].replace(ed_clean_map)
ed_df.head()
```

```python
ed_graph=px.violin(ed_df,
                   x='EdLevel',
                   y='ConvertedCompYearly',
                   color='EdLevel',
                   hover_data=['WorkExp'],
                   box=True,
                   title='Salary vs EdLevel',
                   template='plotly_dark'
                   )
ed_graph.update_layout(xaxis_title='EdLevel',yaxis_title='Salary')
ed_graph.show()
```

## Remote vs. Hybrid vs. On-Site Dynamics

```python

remote_df=df.dropna(subset=["ConvertedCompYearly","RemoteWork"]).copy()
C_remote_df=remote_df[(remote_df["ConvertedCompYearly"]>=5000) & (remote_df["ConvertedCompYearly"]<=800000) & (remote_df['RemoteWork'].str.contains('Remote|Hybrid|In-person|Your choice'))]
remote_df=C_remote_df[["RemoteWork","ConvertedCompYearly"]]
remote_df=remote_df.replace({'RemoteWork': {'Remote': 'Remote', 'Hybrid (some in-person, leans heavy to flexibility)': 'Hybrid_Flexible', 'In-person': 'On-Site','Hybrid (some remote, leans heavy to in-person)':'Hybrid_On-Site','Your choice (very flexible, you can come in when you want or just as needed)':'My choice'}})

remote_df.head()
```

#### Visuals

```python
graph_remote_violin=px.violin(remote_df,
                              x='RemoteWork',
                              y='ConvertedCompYearly',
                              box=True,
                              title='Salary vs Work_Type',
                              template='plotly_dark',
                              color='RemoteWork'
                            )
graph_remote_violin.update_layout(xaxis_title='Work_Type',yaxis_title='Annual_Salary',showlegend=False)
graph_remote_violin.show()
```

#### Popularity OF Work_Type

```python
popu_remote_df=remote_df.groupby('RemoteWork').agg(AvgSalary=('ConvertedCompYearly', 'mean'),
                                                   DeveloperCount=('RemoteWork', 'count')).reset_index().copy()


graph_work_scatter=px.scatter(
                              popu_remote_df,
                              x='DeveloperCount',
                              y='AvgSalary',
                              text='RemoteWork',
                              size='DeveloperCount',
                              color='AvgSalary',
                              color_continuous_scale='plasma',
                              title='Market Dynamics: Work_Type Popularity vs. Average Salary',
                              template='plotly_dark'

)
graph_work_scatter.update_traces(textposition='top center')
graph_work_scatter.update_layout(coloraxis_showscale=False)
graph_work_scatter.show()
```
