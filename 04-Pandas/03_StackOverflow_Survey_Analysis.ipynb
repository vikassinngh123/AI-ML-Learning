## Work exp V/s pay

```python
work_df=df.dropna(subset=["WorkExp","ConvertedCompYearly"])

world_work_df=work_df[(work_df["ConvertedCompYearly"]>0) & (work_df["ConvertedCompYearly"]<1000000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<30)]
cleaned_world_work_df=world_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_world_work_df=cleaned_world_work_df.groupby("WorkExp")['ConvertedCompYearly'].mean().reset_index()

#USA
USA_work_df=work_df[(work_df["ConvertedCompYearly"]>0) & (work_df["ConvertedCompYearly"]<1000000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<30) & (work_df['Country'].isin(['United States of America']))]
cleaned_USA_work_df=USA_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_USA_work_df=cleaned_USA_work_df.groupby("WorkExp")['ConvertedCompYearly'].mean().reset_index()

#India
India_work_df=work_df[(work_df["ConvertedCompYearly"]>0) & (work_df["ConvertedCompYearly"]<500000) & (work_df["WorkExp"]>0) & (work_df["WorkExp"]<30) & (work_df['Country'].isin(['India']))]
cleaned_India_work_df=India_work_df[["WorkExp","ConvertedCompYearly"]]
cleaned_India_work_df=cleaned_India_work_df.groupby("WorkExp")['ConvertedCompYearly'].mean().reset_index()

```

##### Visual

## **Key Insights: Salary vs. Work Experience**

* **The US Market Premium:** The US consistently leads in compensation across all experience levels due to its highly mature tech ecosystem and premium software engineering labor costs.

* **Global Trend Stability:** The 'World' trendline remains incredibly smooth because its massive global sample size completely averages out regional economic extremes.

* **India's High-Experience Volatility**: India’s sharp spikes and dips past 20 years are caused by low sample sizes and extreme outlier sensitivity, a direct byproduct of the country's post-2000s IT boom.

* **The 15-Year Growth Plateau:** Across all regions, aggressive salary growth driven by early-career job-hopping levels off after 15–20 years as developers reach senior management ceilings.


```python
from matplotlib.lines import lineStyles
from matplotlib import markers
plt.figure(figsize=(15,8))
sns.set_theme(style='whitegrid',context='talk')

# World
sns.lineplot(
             data=cleaned_world_work_df,
             x="WorkExp",
             y="ConvertedCompYearly",
             marker='o',
             markersize=10,
             markeredgecolor='black',
             markerfacecolor='white',
             markeredgewidth=2,
             linestyle='-.',
             color='red',
             linewidth=2,
             label='Avg Annual Salary(World)'
             )

# USA
sns.lineplot(
             data=cleaned_USA_work_df,
             x="WorkExp",
             y="ConvertedCompYearly",
             marker='o',
             markersize=10,
             markeredgecolor='black',
             markerfacecolor='white',
             markeredgewidth=2,
             linestyle='-.',
             color='green',
             linewidth=2,
             label='Avg Annual Salary(USA)'
             )

#India
sns.lineplot(
             data=cleaned_India_work_df,
             x="WorkExp",
             y="ConvertedCompYearly",
             marker='o',
             markersize=10,
             markeredgecolor='black',
             markerfacecolor='white',
             markeredgewidth=2,
             linestyle='-.',
             color='blue',
             linewidth=2,
             label='Avg Annual Salary(India)'
             )

plt.title("Salary as work exp increase",fontsize=16,fontweight='bold')
plt.xlabel("Work Exp",fontsize=12,fontweight='bold')
plt.ylabel("Salary",fontsize=12,fontweight='bold')


plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.grid(axis='x',visible=False)

plt.legend(bbox_to_anchor=(1.05,1) ,loc='upper left' ,frameon=False)

sns.despine()

plt.show()
```
